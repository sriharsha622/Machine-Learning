Splitting the country column with unique values and creating the dictinoary with countries as keys and their counts as values 

```python
def unique_column_values(col_values):
  unique = []
  for i in col_values:
    j = i.split(",")
    if len(j)>1:
      for k in j:
        k = k.strip()
        if k not in unique:
          unique.append(k)
    elif j[0].strip() not in unique:
      unique.append(j[0])
  dictt = {key: 0 for key in unique}
  for value in unique:
    for row in col_values:
      j = row.split(",")
      for i in range(len(j)):
        j[i] = j[i].strip()

      if value in j:
        dictt[value]+=1
  return unique,dictt


```

```python
country_unique,country_dict = unique_column_values(df['country'])
print(country_dict)
```

```python
country_dict = dict(sorted(country_dict.items(),key = lambda x: x[1],reverse=True))
print(country_dict)
```

Splitting the listed_in column with unique values and creating the dictinoary with genre as keys and their counts as values

```python
listed_in_unique, listed_in_dict = unique_column_values(df['listed_in'])
print(listed_in_dict)
```

Splitting the cast column with unique values and creating the dictinoary with cast as keys and their counts as values

```python
cast_unique, cast_dict = unique_column_values(df['cast'][1:464])
print(cast_dict)
```
